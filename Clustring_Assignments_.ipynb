# Theoretical Questions

#### 1) What is unsupervised learning in the context of machine learning ?

Unsupervised learning in machine learning involves training algorithms on unlabeled data, allowing them to discover hidden patterns and structures independently.

#### 2) How does K-Means clustering algorithm work ?

K-means clustering, a popular unsupervised machine learning algorithm, works by grouping data points into clusters based on their similarity.

#### 3) Explain the concept of a dendrogram in hierarchical clustering.

A dendrogram is a tree-like diagram that visually represents the results of hierarchical clustering.

#### 4) What is the main difference between K-Means and Hierarchical Clustering ?

The primary difference between K-means and hierarchical clustering lies in how they approach cluster formation: K-means partitions data into a predefined number of non-overlapping clusters, while hierarchical clustering creates a hierarchy of clusters, potentially showing nested relationships.

#### 5) What are the advantages of DBSCAN over K-Means ?

DBSCAN offers several advantages over K-Means, primarily in its ability to handle complex cluster shapes, its robustness to noise, and its flexibility in determining the number of clusters.

#### 6) When would you use Silhouette Score in clustering ?

The Silhouette Score is primarily used to evaluate the quality of clusters, especially when ground truth labels are unavailable, and to determine the optimal number of clusters in a clustering algorithm like K-means.

#### 7)  What are the limitations of Hierarchical Clustering ?

Hierarchical clustering, while powerful, has several limitations. Primarily, it struggles with large datasets due to high time and space complexity, making it computationally expensive.

#### 8) Why is feature scaling important in clustering algorithms like K-Means ?

Feature scaling is crucial in clustering algorithms like K-Means because it ensures all features contribute equally to the clustering process and prevents features with larger ranges from dominating the distance calculations.

#### 9)  How does DBSCAN identify noise points ?

It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster.

#### 10)  Define inertia in the context of K-Means.

It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster.

#### 11) What is the elbow method in K-Means clustering ?

The elbow method is a graphical method for finding the optimal K value in a k-means clustering algorithm. The elbow graph shows the within-cluster-sum-of-square (WCSS) values on the y-axis corresponding to the different values of K (on the x-axis). The optimal K value is the point at which the graph forms an elbow.

#### 12) Describe the concept of "density" in DBSCAN.

In DBSCAN, density refers to the number of data points within a specified radius around a given point. This concept is used to identify clusters as regions of high density separated by areas of low density. DBSCAN classifies data points as core points (dense regions), border points (within a core point's neighborhood), and noise points (isolated or outliers) based on their local density.

#### 13) Can hierarchical clustering be used on categorical data ?

Yes, hierarchical clustering can be used for categorical data, but it requires specific considerations and the use of appropriate distance metrics. Unlike k-means clustering, which is designed for numerical data, hierarchical clustering algorithms like agglomerative clustering can handle categorical data by using distance metrics like Hamming distance or Jaccard similarity.

#### 14) What does a negative Silhouette Score indicate ?

A negative Silhouette Score indicates that a data point has been assigned to the wrong cluster, meaning it's closer to a different cluster than its own. A Silhouette Score ranges from -1 to +1, with positive scores suggesting good clustering, and scores near 0 indicating overlapping clusters.

#### 15) Explain the term "linkage criteria" in hierarchical clustering.

In hierarchical clustering, linkage criteria define how the distance between two clusters is calculated. This determines which clusters are merged at each step of the clustering process. Different linkage criteria, like single linkage, complete linkage, and average linkage, lead to different clustering results, impacting the granularity and geometry of the final clusters.

#### 16) Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

K-Means clustering can struggle with data exhibiting varying cluster sizes or densities because it assumes clusters are spherical and roughly equal in size. This means it can be biased towards creating clusters of similar sizes, even if the true underlying data doesn't align with this assumption. Specifically, larger clusters can disproportionately influence the centroid placement, potentially misclassifying smaller, more dense clusters.

#### 17) What are the core parameters in DBSCAN, and how do they influence clustering ?

DBSCAN's core parameters are Epsilon (Îµ) and MinPts. Epsilon defines the radius within which points are considered neighbors, while MinPts specifies the minimum number of points required within a radius to form a dense region (a core point). These parameters significantly influence the number and shape of identified clusters.

#### 18) How does K-Means++ improve upon standard K-Means initialization ?

K-Means++ improves upon standard K-Means initialization by intelligently selecting initial cluster centroids, aiming to spread them out and avoid clustering into suboptimal solutions. Standard K-Means randomly chooses initial centroids, which can lead to poor cluster assignments, especially when clusters are not well-separated. K-Means++ addresses this by choosing the first centroid randomly and then selecting subsequent centroids based on their distance from previously chosen centroids, effectively pushing them further apart, according to AI Accelerator Institute.

#### 19) What is agglomerative clustering ?

Agglomerative clustering is a hierarchical clustering method that uses a bottom-up approach to group similar data points together. It starts with each data point in its own individual cluster and then iteratively merges the closest clusters until all data points are combined into a single, large cluster. This process creates a hierarchy of clusters, visualized as a dendrogram.

#### 20) What makes Silhouette Score a better metric than just inertia for model evaluation ?

The Silhouette Score is a better metric than just inertia for model evaluation in clustering because it provides a more comprehensive and nuanced assessment of cluster quality by considering both cohesion within clusters and separation between them. Inertia, on the other hand, only focuses on the compactness of clusters, potentially overlooking poorly separated clusters.

#### 21) Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot ?

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualize using a scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5);
plt.show()
```

#### 29) Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k values')
plt.xticks(k_values)
plt.show()
```

#### 31) Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=1)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X)
labels = kmeans.predict(X)

# Plot decision boundaries
# Create a mesh grid
h = 0.05
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict labels for each point in mesh grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the contour and training points
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Accent')
plt.scatter(X[:, 0], X[:, 1], c=labels, edgecolor='k', cmap='Accent')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=1)
labels = kmeans.fit_predict(X)

# Calculate the silhouette coefficient for each sample
silhouette_vals = silhouette_samples(X, labels)

# Plot the silhouette coefficients
plt.figure(figsize=(8, 6))
plt.bar(range(len(silhouette_vals)), silhouette_vals, color='skyblue')
plt.title('Silhouette Coefficient for Each Sample (Iris Dataset)')
plt.xlabel('Sample Index')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

#### 46) Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

#### 47) Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)

# Count the number of clusters and noise points
# Noise points are labeled as -1 by DBSCAN
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Excluding noise
n_noise = list(labels).count(-1)  # Noise points are labeled as -1

# Print the results
print(f"Number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

#### 48) Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load the Digits dataset
digits = load_digits()
X = digits.data  # Features

# Reduce dimensionality with t-SNE (2 components for 2D visualization)
tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=10)  # 10 clusters for 10 digits
labels = agg_clustering.fit_predict(X_tsne)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50)
plt.title('Agglomerative Clustering on Digits Dataset (t-SNE Reduced)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
